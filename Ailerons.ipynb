```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ailerons_train.csv')
```

```python
df
```

```python
# check for null values.

df.isnull().sum()
```

```python
# set the independent and target variable.

x = df.drop('goal',axis=1)
y = df['goal']
```

```python
# remove all columns which have less variance in their data as they won't make meaningful contribution towards the target variable.

from sklearn.feature_selection import VarianceThreshold

const_filter = VarianceThreshold(threshold=0.0001)

x_filter = const_filter.fit_transform(x)
```

```python
# so only eleven columns which have greater than 0.01 % variation in their data.

x_filter = pd.DataFrame(x_filter,columns=x[x.columns[const_filter.get_support(indices=True)]].columns)
x_filter.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
import statsmodels.api as sm
from statsmodels.regression.linear_model import OLS
from sklearn.model_selection import cross_val_score
```

```python
# polynomial function helps to increase the number of independent variables by factoring in the interaction effect among the variables.

from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_trans = PolynomialFeatures(degree=2)

poly_features = poly_trans.fit_transform(x_filter)
```

```python
poly_trans.get_feature_names(input_features=x_filter.columns)
```

```python
interaction_df = pd.DataFrame(poly_features,columns=poly_trans.get_feature_names(input_features=x_filter.columns))
```

```python
interaction_df
```

```python
x_train,x_test,y_train,y_test = train_test_split(interaction_df,y,test_size=0.2)
```

```python
x = sm.add_constant(interaction_df)
lr_ols = OLS(y,x)
results = lr_ols.fit()

# print the summary
print(results.summary())
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
np.mean(cross_val_score(lr,interaction_df,y,scoring='r2',cv=10))
```
